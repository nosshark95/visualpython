```python
import pybinding as pb
import numpy as np
import matplotlib.pyplot as plt

pb.pltutils.use_style()
%matplotlib inline
```

```python
from math import sqrt

def graphene_nanoribbon():
    #a = 0.24595   # [nm] unit cell length
    #a_cc = 0.142  # [nm] carbon-carbon distance
    d = 0.142  # [nm] carbon-carbon distance
    t = -2.8      # [eV] nearest neighbour hopping

    lat = pb.Lattice(a1=[d*5*sqrt(3), 0],a2=[0, 6*d])
    lat.add_sublattices(('0', [0, 0]),
                        ('1', [d*sqrt(3)/2,  d/2]),
                        ('2', [d*sqrt(3)/2,  d*3/2]),
                        ('3', [0,  d*2]),
                        ('4', [0,  d*3]),
                        ('5', [d*sqrt(3)/2,  d*7/2]),
                        ('6', [d*sqrt(3)/2,  d*9/2]),
                        ('7', [0,  d*5]),
                        
                        ('8', [d*sqrt(3), 0]),
                        ('9', [d*3*sqrt(3)/2,  d/2]),
                        ('10', [d*3*sqrt(3)/2,  d*3/2]),
                        ('11', [d*sqrt(3),  d*2]),
                        ('12', [d*sqrt(3),  d*3]),
                        ('13', [d*3*sqrt(3)/2,  d*7/2]),
                        ('14', [d*3*sqrt(3)/2,  d*9/2]),
                        ('15', [d*sqrt(3),  d*5]),
                        
                        ('16', [d*2*sqrt(3), 0]),
                        ('17', [d*5*sqrt(3)/2,  d/2]),
                        ('18', [d*5*sqrt(3)/2,  d*3/2]),
                        ('19', [d*2*sqrt(3),  d*2]),
                        ('20', [d*2*sqrt(3),  d*3]),
                        ('21', [d*5*sqrt(3)/2,  d*7/2]),
                        ('22', [d*5*sqrt(3)/2,  d*9/2]),
                        ('23', [d*2*sqrt(3),  d*5]),
                        
                        ('24', [d*3*sqrt(3), 0]),
                        ('25', [d*7*sqrt(3)/2,  d/2]),
                        ('26', [d*7*sqrt(3)/2,  d*3/2]),
                        ('27', [d*3*sqrt(3),  d*2]),
                        ('28', [d*3*sqrt(3),  d*3]),
                        ('29', [d*7*sqrt(3)/2,  d*7/2]),
                        ('30', [d*7*sqrt(3)/2,  d*9/2]),
                        ('31', [d*3*sqrt(3),  d*5]),
                        
                        ('32', [d*4*sqrt(3), 0]),
                        ('33', [d*9*sqrt(3)/2,  d/2]),
                        ('34', [d*9*sqrt(3)/2,  d*3/2]),
                        ('35', [d*4*sqrt(3),  d*2]),
                        ('36', [d*4*sqrt(3),  d*3]),
                        ('37', [d*9*sqrt(3)/2,  d*7/2]),
                        ('38', [d*9*sqrt(3)/2,  d*9/2]),
                        ('39', [d*4*sqrt(3),  d*5])
                       )
    lat.add_hoppings(
        # inside the main cell
        ([0,  0], '0', '1', t),
        ([0,  0], '1', '2', t),
        ([0,  0], '2', '3', t),
        ([0,  0], '3', '4', t),
        ([0,  0], '4', '5', t),
        ([0,  0], '5', '6', t),
        ([0,  0], '6', '7', t),
        
        ([0,  0], '8', '9', t),
        ([0,  0], '9', '10', t),
        ([0,  0], '10', '11', t),
        ([0,  0], '11', '12', t),
        ([0,  0], '12', '13', t),
        ([0,  0], '13', '14', t),
        ([0,  0], '14', '15', t),
        
        ([0,  0], '16', '17', t),
        ([0,  0], '17', '18', t),
        ([0,  0], '18', '19', t),
        ([0,  0], '19', '20', t),
        ([0,  0], '20', '21', t),
        ([0,  0], '21', '22', t),
        ([0,  0], '22', '23', t),
        
        ([0,  0], '24', '25', t),
        ([0,  0], '25', '26', t),
        ([0,  0], '26', '27', t),
        ([0,  0], '27', '28', t),
        ([0,  0], '28', '29', t),
        ([0,  0], '29', '30', t),
        ([0,  0], '30', '31', t),
        
        ([0,  0], '32', '33', t),
        ([0,  0], '33', '34', t),
        ([0,  0], '34', '35', t),
        ([0,  0], '35', '36', t),
        ([0,  0], '36', '37', t),
        ([0,  0], '37', '38', t),
        ([0,  0], '38', '39', t),
        
        ([0,  0], '1', '8', t),
        ([0,  0], '2', '11', t),
        ([0,  0], '5', '12', t),
        ([0,  0], '6', '15', t),
        
        ([0,  0], '9', '16', t),
        ([0,  0], '10', '19', t),
        ([0,  0], '13', '20', t),
        ([0,  0], '14', '23', t),
        
        ([0,  0], '17', '24', t),
        ([0,  0], '18', '27', t),
        ([0,  0], '21', '28', t),
        ([0,  0], '22', '31', t),
        
        ([0,  0], '25', '32', t),
        ([0,  0], '26', '35', t),
        ([0,  0], '29', '36', t),
        ([0,  0], '30', '39', t),
        # between neighboring cells
        ([-1, 0], '0', '33', t),
        ([-1, 0], '3', '34', t),
        ([-1, 0], '4', '37', t),
        ([-1, 0], '7', '38', t),
        
        ([0, -1], '0', '7', t),
        ([0, -1], '8', '15', t),
        ([0, -1], '16', '23', t),
        ([0, -1], '24', '31', t),
        ([0, -1], '32', '39', t)
    )
    return lat

lattice = graphene_nanoribbon()
lattice.plot()
plt.show()
```

```python
from pybinding.repository import graphene
from math import sqrt, pi
plt.rcParams['figure.figsize']=(30,5)
model = pb.Model(
    graphene_nanoribbon(),
    pb.translational_symmetry(a1=True, a2= True)
)
#print(model.hamiltonian.todense())

plt.subplot(141)
model.plot()
plt.title('Model system')
plt.show()

solver = pb.solver.lapack(model)
#print(solver.eigenvalues)

plt.subplot(142)
bands = solver.calc_bands(-pi/(5*graphene.a), pi/(5*graphene.a)) #in the first Brillouin zone
bands.plot(point_labels=[r"$-\pi / a$", r"$\pi / a$"])
plt.title('Band structure')
plt.show()

plt.subplot(143)
dos = solver.calc_dos(energies=np.linspace(-10, 10, 200), broadening=0.5)  # [eV]
dos.plot()
plt.title('Density of states')
plt.show()

plt.subplot(144)
reference_energy=0
ldos_map = solver.calc_spatial_ldos(energy=reference_energy, broadening=0.05)  # [eV]
ldos_map.plot()
pb.pltutils.colorbar(label='at energy %d'%reference_energy)
plt.title('Local density of states')
plt.show()
```

## Addtion atom on the specific site in unit cell

```python
def addatom(delta_energy):
    #Break sublattice symmetry with opposite A and B onsite energy
    @pb.onsite_energy_modifier
    def potential(energy, sub_id):
        energy[sub_id == '19'] += delta_energy
        return energy
    return potential

model = pb.Model(
    graphene_nanoribbon(),
    pb.translational_symmetry(a1=True, a2= True),
    addatom(1)
)
#print(model3.hamiltonian.todense())

model.onsite_map.plot_contourf()
pb.pltutils.colorbar(label="U (eV)")
model.plot()
plt.title('Model system')
#model.lattice.plot_vectors(position=[0.6, -0.25])
```

## Make vacancies

```python
def vacancy(atom):
    @pb.site_state_modifier
    def modifier(state, sub_id):
        state[sub_id==atom] = False
        return state
    return modifier
model = pb.Model(
    graphene_nanoribbon(),
    pb.translational_symmetry(a1=True, a2= True),
    vacancy('2')
)
#print(model3_v.hamiltonian.todense())

#model3_v.onsite_map.plot_contourf()
#pb.pltutils.colorbar(label="U (eV)")
model.plot()
plt.title('Model system')
```

## Substitution atom in unit cell

```python
def substitute(delta_energy, fix_hopping):
    @pb.onsite_energy_modifier
    def potential(energy, sub_id):
        energy[sub_id == '3'] += delta_energy
        return energy
    @pb.hopping_generator
    def hopping(energy, hop_id):
        energy[hop_id == 't'] = fix_hopping
        return energy, hop_id
    return potential, hopping

model3_h = pb.Model(
    graphene_nanoribbon(),
    pb.translational_symmetry(a1=True, a2= False),
    substitute(2.6,5)
)
print(model3_h.hamiltonian.todense())

model3_h.onsite_map.plot_contourf()
pb.pltutils.colorbar(label="U (eV)")
model3_h.plot()
#model.lattice.plot_vectors(position=[0.6, -0.25])
```
