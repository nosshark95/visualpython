## Addtion an hydrogen atom on the specific site in unit cell

```python
def addatom(delta_energy):
    #Break sublattice symmetry with opposite A and B onsite energy
    @pb.onsite_energy_modifier
    def potential(energy, sub_id):
        energy[sub_id == '55'] += delta_energy
        return energy
    return potential

model = pb.Model(
    graphene_sheet(),
    pb.translational_symmetry(a1=True, a2= True),
    addatom(1)
)
#print(model3.hamiltonian.todense())

model.onsite_map.plot_contourf()
pb.pltutils.colorbar(label="U (eV)")
model.plot()
plt.title('Model system')
#model.lattice.plot_vectors(position=[0.6, -0.25])
```

## Substitution by nitrogen atom, in hexagon shape

```python
def hexagon_substitute(delta_energy, fix_cn_hopping, fix_nn_hopping):
    @pb.onsite_energy_modifier
    def potential(energy, sub_id):
        energy[sub_id == '54'] += delta_energy
        energy[sub_id == '55'] += delta_energy
        energy[sub_id == '56'] += delta_energy
        energy[sub_id == '57'] += delta_energy
        energy[sub_id == '71'] += delta_energy
        energy[sub_id == '72'] += delta_energy
        return energy
    @pb.hopping_energy_modifier
    def hopping(energy, hop_id):
        energy[hop_id == 'cn'] *= fix_cn_hopping
        energy[hop_id == 'nn'] *= fix_cn_hopping
        return energy
    return potential, hopping

model = pb.Model(
    graphene_sheet(),
    pb.translational_symmetry(a1=True, a2= True),
    hexagon_substitute(1,6/7, 5/7)
)
#print(model3.hamiltonian.todense())

model.onsite_map.plot_contourf()
pb.pltutils.colorbar(label="U (eV)")
model.plot()
plt.title('Model system')
#model.lattice.plot_vectors(position=[0.6, -0.25])
```

```python
differ=[]       
for point, val in list(class1.items()):
    i,j=point
    differ.append(abs(i-j))
#print(differ,'\n')

count={}
for difference in differ:
    count[difference]=count.get(difference,0)+1
print(count)
```
