```python
from pybinding.repository import graphene
from math import sqrt, pi
plt.rcParams['figure.figsize']=(30,5)
model = pb.Model(
    graphene_model(),
    pb.translational_symmetry(a1=True, a2= True)
)
#print(model.hamiltonian.todense())

plt.subplot(141)
model.plot()
plt.title('Model system')
plt.close()

solver = pb.solver.lapack(model)
#print(solver.eigenvalues)

plt.subplot(142)
bands = solver.calc_bands(-pi/(7*graphene.a), pi/(7*graphene.a)) #in the first Brillouin zone, for kx
bands.plot(point_labels=[r"$-\pi / a$", r"$\pi / a$"])
plt.title('Band structure')
plt.show()

plt.subplot(143)
dos = solver.calc_dos(energies=np.linspace(-1, 1, 200), broadening=0.05)  # [eV]
dos.plot()
plt.title('Density of states')
plt.show()

plt.subplot(144)
reference_energy=0
ldos_map = solver.calc_spatial_ldos(energy=reference_energy, broadening=0.05)  # [eV]
ldos_map.plot()
pb.pltutils.colorbar(label='at energy %d'%reference_energy)
plt.title('Local density of states')
plt.close()
```

```python
def addatom(delta_energy):
    @pb.onsite_energy_modifier
    def potential(energy, sub_id):
        energy[sub_id == '54'] += delta_energy
        energy[sub_id == '55'] += delta_energy
        energy[sub_id == '56'] += delta_energy
        energy[sub_id == '57'] += delta_energy
        energy[sub_id == '71'] += delta_energy
        energy[sub_id == '72'] += delta_energy
        return energy
    return potential

model = pb.Model(
    graphene_model(),
    pb.translational_symmetry(a1=True, a2= True),
    addatom(3)
)
#print(model3.hamiltonian.todense())

model.onsite_map.plot_contourf()
pb.pltutils.colorbar(label="U (eV)")
model.plot()
plt.title('Model system')
#model.lattice.plot_vectors(position=[0.6, -0.25])
```
